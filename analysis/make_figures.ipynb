```python
# Import needed libraries

import numpy as np
import configparser
import matplotlib.pyplot as plt

def simpleaxis(ax):
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['bottom'].set_visible(False)
    ax.spines['left'].set_visible(False)
    ax.get_xaxis().tick_bottom()
    ax.get_yaxis().tick_left()

semi_model_ont_path = "../results/semi_train_ONT_chemistry.npz"


semi_model_ont = np.load(semi_model_ont_path)

print(list(semi_model_ont.keys()))

num_pos_ont = semi_model_ont['num_pos'] 
total_pos_ont = semi_model_ont['total_pos'] 

print(num_pos_ont)

fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(14, 6))

x_revio = np.arange(0, len(num_pos_ont))
ax.plot(x_revio, num_pos_ont, c="#F41C90", label=f"semi-CNN with ONT model start \n (total positives: {total_pos_ont})")


ax.set_xticks(x_revio)
ax.set_xticklabels(x_revio)

ax.set_ylim([-1000, 50000])

ax.set_xlabel("epochs", fontsize=14)
ax.set_ylabel(f"#m6As at 95% estimated precision \n in the validation set", fontsize=14)


simpleaxis(ax)
ax.grid(True)

plt.legend(bbox_to_anchor=(1.0, 1.0), fontsize=14)
plt.tight_layout()
plt.savefig("../figures/sup_figure_semi_supervised_ont.png", transparent=True)
plt.show()
```
